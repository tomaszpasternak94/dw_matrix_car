```
def group_and_plot(paramX,paramY='price_value',agg_func=[np.median,np.mean,np.size],feat_sort = 'mean',how_much=50,subplots=True):
  return (
     df
    .groupby(paramX)[paramY]
    .agg(agg_func)
    .sort_values(by=feat_sort,ascending=False)
    .head(how_much)
    .plot(kind='bar',figsize=(20,8),subplots=subplots)
    
  )
```

```
group_and_plot('param_marka-pojazdu')
```

```
group_and_plot('param_kolor',feat_sort='mean')
```

```
group_and_plot('param_pojemność-skokowa',feat_sort='size')
```
